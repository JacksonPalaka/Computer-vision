```python
# Play an video 
import cv2

capture = cv2.VideoCapture("Images/mountain.mp4")

if not capture.isOpened():
    print("There is no video ")
    
while True:
    res,frame = capture.read()
    
    if not res:
        break
    
    cv2.imshow("Picture",frame)
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
    
capture.release()
cv2.destroyAllWindows()
```
